```python
## this allow allow to rename all teh images located in a folder with the name of the folder as prefix, for example dill_0.jpg, dill_1.jpg, etc.

from pathlib import Path

folder = Path("/home/jouell/code/jouell3/plant_detect/notebooks/dataset_aromatic_duckduckgo/origan")
prefix = folder.name

for path in sorted(folder.glob("*.jpg")):
    if not path.stem.startswith(f"{prefix}1_"):  # skip already-renamed files
        name = path.name.split("_")
        new_path = path.parent / f"{name[0]}1_{name[-1]}"
        path.rename(new_path)
        print(f"{path.name} → {new_path.name}")

print("Done.")
```
