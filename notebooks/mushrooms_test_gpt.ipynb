```python
import os
import random
import shutil
from pathlib import Path

# ========= PARAMETRES =========
DATASET_DIR = Path("../data/dataset")   # dossier source
OUTPUT_DIR = Path("../data/split_dataset")  # dossier de sortie
SEED = 42

TRAIN_RATIO = 0.70
VAL_RATIO = 0.15
TEST_RATIO = 0.15

random.seed(SEED)

# ========= VERIFICATIONS =========
assert abs(TRAIN_RATIO + VAL_RATIO + TEST_RATIO - 1.0) < 1e-6, "Les ratios doivent faire 1.0"

class_dirs = [d for d in DATASET_DIR.iterdir() if d.is_dir()]
class_dirs = sorted(class_dirs)

print(f"Nombre de classes trouvées : {len(class_dirs)}\n")

class_counts = {}

for class_dir in class_dirs:
    images = [f for f in class_dir.iterdir() if f.is_file()]
    class_counts[class_dir.name] = len(images)
    print(f"{class_dir.name}: {len(images)} images")

# ========= CREATION DOSSIERS =========
for split in ["train", "val", "test"]:
    for class_name in class_counts.keys():
        (OUTPUT_DIR / split / class_name).mkdir(parents=True, exist_ok=True)

# ========= SPLIT =========
for class_dir in class_dirs:
    class_name = class_dir.name
    images = [f for f in class_dir.iterdir() if f.is_file()]
    random.shuffle(images)

    n_total = len(images)
    n_train = int(n_total * TRAIN_RATIO)
    n_val = int(n_total * VAL_RATIO)

    train_files = images[:n_train]
    val_files = images[n_train:n_train + n_val]
    test_files = images[n_train + n_val:]

    for file_path in train_files:
        shutil.copy(file_path, OUTPUT_DIR / "train" / class_name / file_path.name)

    for file_path in val_files:
        shutil.copy(file_path, OUTPUT_DIR / "val" / class_name / file_path.name)

    for file_path in test_files:
        shutil.copy(file_path, OUTPUT_DIR / "test" / class_name / file_path.name)

    print(f"\nClasse: {class_name}")
    print(f"  Train: {len(train_files)}")
    print(f"  Val  : {len(val_files)}")
    print(f"  Test : {len(test_files)}")

print("\nSplit terminé.")
```

```python
import tensorflow as tf
from pathlib import Path

# ========= PARAMETRES =========
DATA_DIR = Path("../data/split_dataset")
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42

# ========= DATASETS =========
train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR / "train",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR / "val",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR / "test",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
num_classes = len(class_names)

print("Classes :", class_names)
print("Nombre de classes :", num_classes)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)
test_ds = test_ds.prefetch(AUTOTUNE)
```

```python
for images, labels in train_ds.take(1):
    print("Shape images :", images.shape)
    print("Shape labels :", labels.shape)
    print("Exemple labels :", labels[:10].numpy())
```

```python
import tensorflow as tf

IMG_SIZE = (224, 224)
NUM_CLASSES = len(class_names)

base_model = tf.keras.applications.EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(224, 224, 3)
)

base_model.trainable = False

inputs = tf.keras.Input(shape=(224, 224, 3))

x = tf.keras.applications.efficientnet.preprocess_input(inputs)
x = base_model(x, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = tf.keras.layers.Dense(NUM_CLASSES, activation="softmax")(x)

model = tf.keras.Model(inputs, outputs)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
EPOCHS = 10

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)
```

```python
val_loss, val_accuracy = model.evaluate(val_ds)

print("Validation loss :", val_loss)
print("Validation accuracy :", val_accuracy)
```

```python
test_loss, test_accuracy = model.evaluate(test_ds)

print("Test loss :", test_loss)
print("Test accuracy :", test_accuracy)
```

```python
print("Dernière train accuracy :", history.history["accuracy"][-1])
print("Dernière val accuracy   :", history.history["val_accuracy"][-1])

print("Dernière train loss :", history.history["loss"][-1])
print("Dernière val loss   :", history.history["val_loss"][-1])
```

```python
print("\n=== Résultats fin d'entraînement ===")
print(f"Train accuracy : {history.history['accuracy'][-1]:.4f}")
print(f"Val accuracy   : {history.history['val_accuracy'][-1]:.4f}")
print(f"Train loss     : {history.history['loss'][-1]:.4f}")
print(f"Val loss       : {history.history['val_loss'][-1]:.4f}")

print("\n=== Évaluation sur le test set ===")
test_loss, test_accuracy = model.evaluate(test_ds)
print(f"Test accuracy  : {test_accuracy:.4f}")
print(f"Test loss      : {test_loss:.4f}")
```
